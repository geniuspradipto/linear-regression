```python
X = df['YearsExperience'].values
y = df['Salary'].values
```

```python
plt.scatter(X,y, c='r', s=10)
plt.grid()
plt.xlabel('YearsExperience -->',fontdict={'family':'serif', 'color':'blue', 'size': 15})
plt.ylabel('Salary -->', fontdict={'family':'serif', 'color':'blue', 'size': 15})
plt.title("Years of Experience VS Salary", fontdict={'family':'serif', 'color':'blue', 'size': 20})
plt.show()
```

#**Linear Regression** - Machine Learning Algorithm, for Regression Analysis

```python
X
```

```python
n = len(X)
X = X.reshape((n,1)) # reshape - from 1D array to matrix convert
X
```

```python
# predictive analysis
from sklearn.linear_model import LinearRegression
model = LinearRegression() # model create
model.fit(X, y) # model train
```

```python
y_pred = model.predict(X)
```

```python
plt.plot(y_pred,X, c='k') # regression line
plt.scatter(y,X, c='r', s=10)
plt.grid()
plt.ylabel('YearsExperience -->',fontdict={'family':'serif', 'color':'blue', 'size': 15})
plt.xlabel('Salary -->', fontdict={'family':'serif', 'color':'blue', 'size': 15})
plt.title("Years of Experience VS Salary", fontdict={'family':'serif', 'color':'blue', 'size': 20})
plt.show()
```

```python
res = model.score(X,y)
print(f"The accuracy of the Linear Regression Model is : {res * 100}%")
```
